```python
# to get an idea about the number of fields present there, find out all the fields present over there

n_list = driver.find_elements(By.XPATH, '//select[@name="pmrId"]')
name_list = []
for i in n_list:
    name_list = i.text.split("\n")
len_name = len(name_list)
print(len_name) #this is the length of the fields present over there
# print(new_list[112])
# print(new_list)

#selecting year
year_ = input("Please enter a year above from 2020(Not 2020): ")
year = driver.find_element(By.NAME,"year")
yr_ele = Select(year)
yr_ele.select_by_visible_text(year_)

#selecting month
month_ = input("Enter a Month name as it is from following- January, February, March, April, May, June, July, August, September, October, November, and December: ")
month = driver.find_element(By.NAME,"month")
mn_ele = Select(month)
mn_ele.select_by_visible_text(month_)
```

```python
#For loop for the number of elements in the dropdown section

for num in range(1,len_name):
    name = driver.find_element(By.NAME,"pmrId")
    name_ele = Select(name)
    name_ele.select_by_visible_text(name_list[num])

    go = driver.find_element(By.XPATH,'//div[@class="go_area go-area"]')
    # print(go.text)
    go.click()
    time.sleep(1)
    
    # Get the page source after clicking the button
    html = driver.page_source

    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    #Finding table and extract data from it 
    try:
        tab1 = soup.find_all("table")[0]
        data1 = tab1.find_all("td")
        data1_ = []
        for i in data1:
            data1_.append(i.text.strip())
        df.loc[num] = data1_
    except Exception:
        df.loc[num] = [0] * len(df.columns)
    # time.sleep(1)

    try:
        tab2 = soup.find_all("table")[1]
        data2 = tab2.find_all("td")
        data2_ = []
        for i in data2:
            data2_.append(i.text.strip())
        df2.loc[num] = data2_[1:8] + data2_[9:16]
    except Exception:
        df2.loc[num] = [0] * len(df2.columns)
    # time.sleep(1)    
    
    try:
        tab3 = soup.find_all("table")[2]
        data3 = tab3.find_all("td")
        data3_ = []
        for i in data3:
            data3_.append(i.text.strip())
        n_data3 = data3_[1:13] + data3_[14:26]
        n_data3.append(data3_[-1])
        df3.loc[num] = n_data3
    except Exception:
        df3.loc[num] = [0] * len(df3.columns)
    # time.sleep(1)
    
    try:
        tab4 = soup.find_all("table")[4]
        data4 = tab4.find_all("td")
        data4_ = []
        for i in data4:
            data4_.append(i.text.strip())
        df4.loc[num] = data4_[1:8] + data4_[9:16]
    except Exception:
        df4.loc[num] = [0] * len(df4.columns)
    # time.sleep(1)
    
    try:
        tab5 = soup.find_all("table")[9]
        data5 = tab5.find_all("td")
        data5_ = []
        for i in data5:
            data5_.append(i.text.strip())
        df5.loc[num] = data5_[1:8] + data5_[9:16]
    except Exception:
        df5.loc[num] = [0] * len(df5.columns)
    # time.sleep(1)
    
    try:
        tab6 = soup.find_all("table")[10]
        data6 = tab6.find_all("th")
        data6_ = []
        for i in data6:
            data6_.append(i.text.strip())
        df6.loc[num] = data6_[-12:]
    except Exception:
        df6.loc[num] = [0] * len(df6.columns)
    # time.sleep(1)
        
    try:
        tab7 = soup.find_all("table")[11]
        data7 = tab7.find_all("td")
        data7_ = []
        for i in data7:
            data7_.append(i.text.strip())
        df7.loc[num] = data7_[-6:]
    except Exception:
        df7.loc[num] = [0] * len(df7.columns)
    # time.sleep(1)

    try:
        tab8 = soup.find_all("table")[14]
        data8 = tab8.find_all("td")
        data8_ = []
        for i in data8:
            data8_.append(i.text.strip())
        df8.loc[num] = data8_[1:8] + data8_[9:16]
    except Exception:
        df8.loc[num] = [0] * len(df8.columns)
    time.sleep(1)
```
